```python
import jieba
text = "小明毕业于北京大学计算机系"
word_generator = jieba.cut(text) # cut方法返回一个生成器
for word in word_generator:
    print(word)
    
word_list = jieba.lcut(text) # lcut方法返回一个列表
print(word_list)
# ['小明', '毕业', '于', '北京大学', '计算机系']
```

```python
import torch
import torch.nn as nn
from gensim.models import KeyedVectors
# 1. 加载预训练词向量
word_vectors = KeyedVectors.load_word2vec_format('./data/my_vectors.kv')
# 2. 构建词表和词向量矩阵
word2index = word_vectors.key_to_index   # 词到索引的映射
embedding_dim = word_vectors.vector_size # 词向量维度
num_embeddings = len(word2index)         # 词表大小
embedding_matrix = torch.zeros((num_embeddings, embedding_dim)) # 初始化词向量矩阵
for word, idx in word2index.items():
    embedding_matrix[idx] = torch.tensor(word_vectors[word]) # 将预训练词向量填入矩阵
# 3. 构建PyTorch的Embedding层
embedding_layer = nn.Embedding.from_pretrained(
    embedding_matrix,  # 预训练词向量矩阵
    freeze=False        # 是否冻结词向量，True表示不更新词向量
)
# 4. 示例：将词索引转换为向量
input_words = ["我", "喜欢", "乘坐", "地铁"] # 分词后的句子
input_indices = [word2index[word] for word in input_words] # 将词转换为索引
input_tensor = torch.tensor(input_indices) # 转换为PyTorch张量
# 5. 查询嵌入
output = embedding_layer(input_tensor) # 获取词向量
print(output) # 输出词向量张量
print(output.shape) # 输出词向量的形状 torch.Size([4, 100])
```

```python
# 处理OOV问题
unk_token = '<UNK>'
index2word = [unk_token] + word_vectors.index_to_key
word2index = {word: idx for idx, word in enumerate(index2word)}
embedding_dim = word_vectors.vector_size # 词向量维度
num_embeddings = len(word2index)         # 词表大小
embedding_matrix = torch.zeros((num_embeddings, embedding_dim)) # 初始化词向量矩阵
for word, idx in word2index.items():
    if word in word_vectors:
        embedding_matrix[idx] = torch.tensor(word_vectors[word]) # 将预训练词向量填入矩阵
    else:
        embedding_matrix[idx] = torch.randn(embedding_dim) * 0.01 # 对于OOV词，使用随机初始化的向量
# 3. 构建PyTorch的Embedding层
embedding_layer = nn.Embedding.from_pretrained(
    embedding_matrix,  # 预训练词向量矩阵
    freeze=False        # 是否冻结词向量，True表示不更新词向量
)
# 4. 示例：将词索引转换为向量
input_words = ["我", "喜欢", "乘坐", "宇宙飞船"] # 分词后的句子
input_indices = [word2index.get(word, 0) for word in input_words] # 将词转换为索引
input_tensor = torch.tensor(input_indices) # 转换为PyTorch张量
# 5. 查询嵌入
output = embedding_layer(input_tensor) # 获取词向量
print(output) # 输出词向量张量
print(output.shape) # 输出词向量的形状 torch.Size([4, 100])
```
