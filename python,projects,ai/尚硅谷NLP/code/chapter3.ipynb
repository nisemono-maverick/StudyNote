```python
import jieba
text = "小明毕业于北京大学计算机系"
word_generator = jieba.cut(text) # cut方法返回一个生成器
for word in word_generator:
    print(word)
    
word_list = jieba.lcut(text) # lcut方法返回一个列表
print(word_list)
# ['小明', '毕业', '于', '北京大学', '计算机系']
```

```python
import jieba
text = "小明毕业于北京大学计算机系"
word_generator = jieba.cut(text, cut_all=True) # cut方法返回一个生成器
for word in word_generator:
    print(word)
    
word_list = jieba.lcut(text, cut_all=True) # lcut方法返回一个列表
print(word_list)
# ['小', '明', '毕业', '于', '北京', '北京大学', '大学', '计算', '计算机', '计算机系', '算机', '系']
```

```python
import jieba
text = "小明毕业于北京大学计算机系"
word_generator = jieba.cut_for_search(text) # cut方法返回一个生成器
for word in word_generator:
    print(word)
    
word_list = jieba.lcut_for_search(text) # lcut方法返回一个列表
print(word_list)
# ['小明', '毕业', '于', '北京', '大学', '北京大学', '计算', '算机', '计算机', '计算机系']
```
